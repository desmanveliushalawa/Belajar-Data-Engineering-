## 1. Navigasi Folder di Git Bash
- **Masuk ke drive D:**
  ```bash
  cd /d/
  ```
- **Cek isi drive D:**
  ```bash
  ls
  ```
- **Masuk ke folder project (contoh: BELAJAR):**
  ```bash
  cd BELAJAR
  ```
- **Jika folder ada spasi (contoh: Belajar Data Engineer):**
  ```bash
  cd "Belajar Data Engineer"
  ```
  atau
  ```bash
  cd Belajar\ Data\ Engineer
  ```
- **Cek posisi sekarang:**
  ```bash
  pwd
  ```
- **Cek isi folder:**
  ```bash
  ls
  ```

## 2. Inisialisasi Git

```bash
git init
```

---

## 3. Menambahkan File ke Staging
- **Tambahkan semua file:**
  ```bash
  git add .
  ```
- **Tambahkan hanya satu file:**
  ```bash
  git add nama_file.py
  ```

---

## 4. Membatalkan Perintah yang Terlanjur
- **Batalkan semua file dari staging:**
  ```bash
  git reset
  ```
- **Batalkan hanya satu file tertentu:**
  ```bash
  git reset nama_file.py
  ```
- **Cek status setelah reset:**
  ```bash
  git status
  ```

---

## 5. Commit Perubahan
```bash
git commit -m "pesan commit pertama"
```

---

## 6. Hubungkan ke Repository GitHub
- Buat repository baru di GitHub (misalnya `Belajar-DataEngineer`).
- Copy URL repository, contoh:
  ```
  https://github.com/username/Belajar-DataEngineer.git
  ```
- Tambahkan remote:
  ```bash
  git remote add origin https://github.com/username/Belajar-DataEngineer.git
  ```
- Cek remote:
  ```bash
  git remote -v
  ```

---

## 7. Push ke GitHub
```bash
git branch -M main
git push -u origin main
```

---

## 8. Tips Troubleshooting
- **Error â€œtoo many argumentsâ€ saat cd:** gunakan tanda kutip atau backslash untuk folder dengan spasi.
- **Error remote origin tidak ada:** tambahkan remote dengan `git remote add origin <URL>`.
- **Sudah salah add file:** gunakan `git reset` atau `git reset nama_file` untuk membatalkan.
- **Cek status kapan saja:**  
  ```bash
  git status
  ```

## Artinya
- Repository lokal dan remote sekarang **sudah terhubung dengan benar**.  
- Kamu bisa lanjut bekerja di folder lokal, lalu setiap kali ada perubahan file:  
  1. `git add nama_file`  
  2. `git commit -m "pesan commit"`  
  3. `git push`  

Semua perubahan akan langsung masuk ke GitHub.
