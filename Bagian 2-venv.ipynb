# VIRTUAL ENVIRONMENT (VENV)

## Panduan Lengkap & Runtut untuk Data Engineer

---

## 1Ô∏è‚É£ Pengertian Virtual Environment

**Virtual Environment (venv)** adalah **lingkungan Python terisolasi** untuk sebuah project.

Artinya:

* setiap project punya:

  * Python environment sendiri
  * daftar library sendiri
  * versi library sendiri

üìå **Isolasi ini WAJIB di dunia profesional**

---

## 2Ô∏è‚É£ Kenapa VENV Itu Wajib?

### Masalah Tanpa VENV

Jika tidak pakai venv:

* semua project pakai library global
* versi bentrok
* project lama rusak saat install library baru

Contoh:

* Project A butuh `pandas==1.5`
* Project B butuh `pandas==2.0`
* tanpa venv ‚Üí chaos üí•

### Dengan VENV

* tiap project aman
* tidak saling ganggu
* reproducible (bisa dijalankan ulang di laptop lain)

üìå **Rule industri**

> Satu project = satu virtual environment

---

## 3Ô∏è‚É£ Kapan VENV Dibuat?

**Urutan yang BENAR:**

1. Buat folder project
2. Init git
3. Commit struktur awal
4. **Baru buat venv**

üìå Alasan:

* venv **tidak boleh ikut git**
* venv itu lokal environment

---

## 4Ô∏è‚É£ Tool yang Digunakan

| Tool    | Fungsi              |
| ------- | ------------------- |
| Python  | Bahasa utama        |
| `venv`  | Modul bawaan Python |
| pip     | Install library     |
| VS Code | Editor + terminal   |
| Git     | Version control     |

Tidak perlu conda / poetry dulu.
**venv = standar industri dasar**

## 6Ô∏è‚É£ Cara Membuat Virtual Environment

Pastikan posisi terminal di root project:

```bash
python -m venv venv
```

Setelah itu muncul folder:

```text
venv/
```

## 7Ô∏è‚É£ Cara Aktivasi VENV

### Windows (PowerShell / CMD)

```bash
venv\Scripts\activate
```

Jika berhasil:

```text
(venv) D:\BELAJAR\project_data_engineer>
```

üìå Tanda `(venv)` = environment aktif

## 9Ô∏è‚É£ requirements.txt

### Apa Itu?

File yang berisi **daftar semua library & versinya**.

### Cara Membuat:

```bash
pip freeze > requirements.txt
```

Contoh isi:

```text
numpy==2.4.2
pandas==3.0.0
python-dotenv==1.2.1
```

---

## üîü Fungsi requirements.txt

1. Reproducible environment
2. Bisa dipakai orang lain
3. Dipakai di server / Docker / CI-CD

üìå Tanpa `requirements.txt`:

* orang lain tidak tahu library apa yang dibutuhkan
* project sulit dijalankan ulang

---

## 1Ô∏è‚É£1Ô∏è‚É£ Install dari requirements.txt

Jika clone repo di laptop baru:

```bash
python -m venv venv
venv\Scripts\activate
pip install -r requirements.txt
```

üî• Environment langsung sama persis
