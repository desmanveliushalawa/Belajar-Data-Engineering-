## Latihan

Buat program:
Baca file CSV
Filter data tertentu
Simpan hasil ke CSV baru

```python
# Baca file CSV

import pandas as pd
df=pd.read_csv("sales.csv")
df.head()

# filter data tertentu, disini saya akan filter sales lebih dari 1000000

filtered_data = df[df["Sales"]>1000000]
print (filtered_data)
```

```python
# simpan hasil filter ke file baru
filtered_data.to_csv("hasil_filter.csv", index = False)
print("data telah berhasil di filter dan disimpan ke file hasil_filter.csv")
```
